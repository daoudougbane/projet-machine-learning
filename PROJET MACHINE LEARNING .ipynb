# ETAPE 2:  Définition du problème

```python
# Objectif : prédire si un passager du Titanic a survécu ou non.
# C'est un problème de classification binaire :
# - 0 = le passager est décédé
# - 1 = le passager a survécu
#- Type : Classification binaire.
# - Contraintes pédagogiques : Pas de régression, utiliser uniquement des modèles de classification.
#
# La variable cible est donc "Survived".
# L'objectif métier est de comprendre quels facteurs (âge, sexe, classe sociale, etc.)
# influencent le plus la survie.

#==========================
# Description des variables 
#==========================

#PassengerId : Identifiant unique du passager.
#Survived : Indique si le passager a survécu (1) ou non (0).
#Pclass : Classe du billet (1 = première classe, 2 = deuxième, 3 = troisième).
#SibSp : Nombre de frères/sœurs ou conjoint à bord.
#Parch : Nombre de parents ou enfants à bord.
#Ticket : Numéro du billet. Peu exploitable brut, mais peut être transformé.
#Fare : Tarif payé pour le billet.
#Cabin : Numéro de cabine. Beaucoup de valeurs manquantes, mais peut donner la position sur le bateau.
#Embarked : Port d’embarquement (C = Cherbourg, Q = Queenstown, S = Southampton).

```

# ETAPE 3 : Prétraitement des données (Data Preprocessing)

```python
import numpy as np
# Définir une fonction pour détecter et remplacer les valeurs aberrantes
def remplacer_aberrantes(df, colonne, seuil_inf=0.1, seuil_sup=0.9):
    q1 = df[colonne].quantile(seuil_inf)
    q3 = df[colonne].quantile(seuil_sup)
    iqr = q3 - q1
    borne_inf = q1 - 1.5 * iqr
    borne_sup = q3 + 1.5 * iqr
    mode = df[colonne].mode()[0]  # Le mode peut être plusieurs valeurs, on prend la première
    df[colonne] = np.where((df[colonne] < borne_inf) | (df[colonne] > borne_sup), mode, df[colonne])
    return df

# Imputation des valeurs aberrantes par le mode
df = remplacer_aberrantes(df, 'Age')

df.isnull().sum()
```

```python
# IMPORTATIONS DES LIBRARIES 
from sklearn.preprocessing import LabelEncoder


# =======================
# Gestion des valeurs manquantes
# =======================
df["Age"].fillna(df["Age"].median(), inplace=True)       # Remplace âge manquant par la médiane
df["Fare"].fillna(df["Fare"].median(), inplace=True)     # Remplace tarif manquant par la médiane
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True) # Remplace port manquant par le plus fréquent

# Supprime la colonne Cabin (trop de valeurs manquantes)
df.drop(columns=['Cabin'], inplace=True, errors='ignore')

# =======================
# Création d'une nouvelle variable "Categorie"
# =======================
df["Categorie"] = pd.cut(df["Age"], 
                         bins=[0,12,18,40,60,80], 
                         labels=["Enfant","Adolescent","Adulte","Senior","Vieux"])

# =======================
# One-Hot Encoding (variables catégorielles)
# =======================
df = pd.get_dummies(df, columns=["Sex","Embarked","Categorie"], drop_first=False)

# =======================
# Vérification des colonnes finales
# =======================
print(df.columns.tolist())

# Visualiser rapidement les valeurs manquantes
sns.heatmap(df.isnull(), cbar=False)
plt.title("Valeurs manquantes dans le dataset")
plt.show()
```

# ETAPE 4: Séparation des données

```python

from sklearn.model_selection import train_test_split

# =======================
# Sélection des variables pertinentes
# =======================
features = ['SibSp', 'Parch', 'Fare','Age','Pclass','Embarked_C','Embarked_Q','Embarked_S','Sex_male','Sex_female','Categorie_Enfant','Categorie_Adulte','Categorie_Adolescent','Categorie_Senior','Categorie_Vieux']
X = df[features]   # Variables explicatives
y = df['Survived'] # Variable cible

# =======================
# Division en train/test
# =======================
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,         # 20% pour le test
    random_state=42,       # Pour garder la même division
    stratify=y             # Conserver la proportion de 0/1
)

print("Taille jeu d’entraînement :", X_train.shape)
print("Taille jeu de test :", X_test.shape)
```

# ETAPE 5: Choix des modèles

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier

# LogitBoost = approximation avec AdaBoost + arbre faible (stump)
logitboost = AdaBoostClassifier(
    base_estimator=DecisionTreeClassifier(max_depth=1, random_state=42),
    n_estimators=300,
    learning_rate=0.2,
    algorithm="SAMME.R",
    random_state=42
)

# Dictionnaire de modèles
models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "AdaBoost": AdaBoostClassifier(n_estimators=100, random_state=42),
    "LogitBoost (approx)": logitboost,
    "Gradient Boosting": GradientBoostingClassifier(
        n_estimators=200, learning_rate=0.1, max_depth=3, random_state=42
    )
}

# Ajout de XGBoost si dispo
try:
    from xgboost import XGBClassifier
    models["XGBoost"] = XGBClassifier(
        n_estimators=300, learning_rate=0.1, max_depth=3,
        subsample=0.9, colsample_bytree=0.9, random_state=42,
        eval_metric="logloss", use_label_encoder=False
    )
except:
    print("⚠ XGBoost non disponible.")
```

# ETAPE 6 :  Entraînement du modèle

```python
# =======================
for name, model in models.items():
    model.fit(X_train, y_train)  # Apprentissage sur les données d’entraînement
    print(f"✅ {name} entraîné avec succès")
```
